```python
import cv2

# 1. 讀取圖片 (預設彩色 BGR)
img_color = cv2.imread("Lenna.jpg", cv2.IMREAD_COLOR)

# 2. 讀取圖片 (灰階)
img_gray = cv2.imread("Lenna.jpg", cv2.IMREAD_GRAYSCALE)

# 3. 讀取圖片 (含透明度，如果有 alpha 通道)
img_unchanged = cv2.imread("Lenna.jpg", cv2.IMREAD_UNCHANGED)

# --- 顯示圖片 ---
cv2.imshow("Color Image", img_color)
cv2.imshow("Gray Image", img_gray)

# 等待任意鍵後關閉視窗
cv2.waitKey(0)
cv2.destroyAllWindows()


cv2.imwrite("output_gray.png", img_gray)
```

```python
import cv2
import matplotlib.pyplot as plt

# 讀取圖片 (BGR 格式)
img = cv2.imread("Lenna.jpg")

# 拆分三個 channel
B = img[:,:,0]
G = img[:,:,1]
R = img[:,:,2]

# 顯示原圖與三個 channel
plt.figure(figsize=(10,5))

plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original (RGB)")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(R)
plt.title("Red channel")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(G)
plt.title("Green channel")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(B)
plt.title("Blue channel")
plt.axis("off")

plt.show()
```

```python
import numpy as np

# 建立全 0 矩陣
zeros = np.zeros_like(R)

# 保留單一通道 
only_R = cv2.merge([zeros, zeros, R])  # 紅色
only_G = cv2.merge([zeros, G, zeros])  # 綠色
only_B = cv2.merge([B, zeros, zeros])  # 藍色

# 顯示
plt.figure(figsize=(10,5))
plt.subplot(1,4,1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.title("Original"); plt.axis("off")
plt.subplot(1,4,2); plt.imshow(cv2.cvtColor(only_R, cv2.COLOR_BGR2RGB)); plt.title("Red channel"); plt.axis("off")
plt.subplot(1,4,3); plt.imshow(cv2.cvtColor(only_G, cv2.COLOR_BGR2RGB)); plt.title("Green channel"); plt.axis("off")
plt.subplot(1,4,4); plt.imshow(cv2.cvtColor(only_B, cv2.COLOR_BGR2RGB)); plt.title("Blue channel"); plt.axis("off")
plt.show()
```
